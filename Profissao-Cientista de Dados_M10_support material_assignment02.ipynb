````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
````

````python
# Carregando a base de dados
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop(['Unnamed: 0', 'index'], axis=1)
df.head()
````

````python
# Verificando a correlação
df.corr()
````

````python
df.info()
````

````python
sns.set(style="darkgrid")
````

````python
# Plot posse de posse de veiculo X renda
ax = sns.pointplot(
     data=df,
     x='posse_de_veiculo',
     y= 'renda',
     dodge=True,
     ci=95
)
plt.title('Posse de Veículo X Renda')
````

````python
# Plot posse de imóvel X reda
ax = sns.pointplot(
    data=df,
    x='posse_de_imovel',
    y='renda',
    # hue='mau',
    dodge=True,
    ci=95
)
plt.title('Posse de imóvel X Renda')
````

````python
# Cria uma nova coluna que também analisará caso o cliente possua ambos imóveis e veículo
df['posse_ambos'] = np.where((df['posse_de_veiculo'] == True) & (df['posse_de_imovel'] == True),
                                    'posse_ambos', 
                                    np.where((df['posse_de_veiculo'] == True) & (df['posse_de_imovel'] == False),
                                    'apenas_veiculo', 
                                    np.where((df['posse_de_veiculo'] == False) & (df['posse_de_imovel'] == True),
                                    'apenas_imovel', 
                                    'nao_possui')))
````

````python
# Plot Possui Imovel e Veículo x Renda
ax = sns.pointplot(x="posse_ambos", y="renda", data=df, dodge=True, ci = 95)
plt.title('Posse de Imóvel e Veículo x Renda')
````

## Análise

Podemos observar nos gráficos que tanto posse de veículo como posse de imóvel são boas preditodas para variável renda. Sendo que posse de imóvel é estatiscamente menos significante.

Na variável posse de ambos combinanos as duas variáveis, posse de veículos e posse de imóvel, nela encotramos uma variável que é uma boa preditora pra variável renda e estatiscamente mais significante que as variáveis posse de veiculo e posse de imóvel analisada separadamente.
