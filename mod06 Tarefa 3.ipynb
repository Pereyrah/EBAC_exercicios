#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc_ro.loc[(sinasc_ro['APGAR5'] >= 0) & (sinasc_ro['APGAR5']<= 3), 'categoria'] = 'asfixia grave'
sinasc_ro.loc[(sinasc_ro['APGAR5'] >= 4) & (sinasc_ro['APGAR5']<= 5),  'categoria'] = 'asfixia moderada'
sinasc_ro.loc[(sinasc_ro['APGAR5'] >= 6) & (sinasc_ro['APGAR5']<= 7), 'categoria'] = 'asfixia leve'
sinasc_ro.loc[(sinasc_ro['APGAR5'] >= 8) & (sinasc_ro['APGAR5']<= 10), 'categoria'] = 'normal'
```

```python
sinasc_ro['categoria'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Tamanho da figura
plt.figure(figsize=(10, 8))
```

```python
#Barplot
sns.catplot(x ='GESTACAO',
            y = 'APGAR5',
            hue= 'SEXO',
            kind='bar',
            data= sinasc_ro,
            ci=False,
            alpha = 0.7,
            height = 8,
            edgecolor = 'black')
# A instância do objeto Axes que está sendo exibida na figura atual
ax=plt.gca()
#Adiciona valores de altura das barras.
for p in ax.patches:
    ax.annotate(f'\n{round(p.get_height(), 2)}', (p.get_x()+0.2, p.get_height()), ha = 'center', color = 'black', size = 10)

# Aumentando a fonte do rótulo do eixo x
for tick in plt.gca().get_xticklabels():
    tick.set_fontsize(9)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab( sinasc_ro['GESTACAO'], sinasc_ro['APGAR5'])
freq
```

```python
#Aqui com as categorias do APGAR

freq = pd.crosstab(sinasc_ro['GESTACAO'], sinasc_ro['categoria'])

freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis=1), axis=0)
freq_pctlin
```

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas. Fique à vontade para escolher a melhor maneira na sua opinião, ou até testar mais de um dos gráficos que vimos.
```python
sns.boxplot(data = sinasc_ro,
           x = 'PESO',
           y = 'categoria',
           order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave'],
           orient = 'h',
           palette = 'husl')
```

. Podemos observar que quanto menor o peso do bebê, maior é a incidência de "asfixia grave" e "asfixia moderada". Porém, isso não quer dizer que esses graus de asfixia aconteçam apenas em bebês com pouco peso.

```python
sns.displot(data = sinasc_ro[sinasc_ro['PESO'] <= 2200]
            , x = 'PESO'
            , hue = 'categoria'
            , hue_order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']
            , bins = 30
            , element = 'step'
            , height = 15
           )
```

. Nesse gráfico acima foram escoclhidos registros que apresentavam o peso menor que 2200g, pois ao utilizar abaixo de 3000g ou 3kg ficava um pouco dificil de analisar 

```python
listagestacao = [['Menos de 22 semanas', '22 a 27 semanas'], 
                 ['28 a 31 semanas', '32 a 36 semanas'], 
                 ['37 a 41 semanas', '42 semanas e mais']]

plt.rc('figure', figsize = (20, 20))

fig, axes = plt.subplots(3, 2)

for i in range(len(listagestacao)):
    for j in range(len(listagestacao[i])):
        sns.histplot(ax = axes[i, j]
                     , x = 'PESO'
                     , hue = 'categoria'
                     , hue_order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']
                     , element = 'step'
                     , data = sinasc_ro[(sinasc_ro['GESTACAO'] == listagestacao[i][j]) 
                                        & (sinasc_ro['PESO'] < 3000)]
                    ).set_title(listagestacao[i][j])
```

. Ao analisar o tempo de gestação, quando antes de 31 semanas, a probabilidade de ter uma asfixia, também estar relacionada com o peso abaixo de 3kg.
